Find top 10 popular car brand and models from 2010 year

```python
def top_popular_car_brand_and_model():
    df['date'] = pd.to_datetime(df['D_REG'], format='%d.%m.%Y')
    df['year'] = pd.DatetimeIndex(df['date']).year
    df['month'] = pd.DatetimeIndex(df['date']).month
    filter_by_year = df[df['year'] >= 2010]
    return filter_by_year.drop_duplicates(subset=['VIN'], keep='last').groupby(['BRAND','MODEL']).size().nlargest(10)
```

```python
top_popular_car_brand_and_model()
```

Find top 10 popular car brand with percentage count

```python
def top_popular_car_brand_with_relative_count():
    total = df['REG_ADDR_KOATUU'].count()
    df_brand = df.groupby(['BRAND']).size().nlargest(10).reset_index(name="count")
    df_brand['%'] = df_brand['count']/total*100
    return df_brand
```

```python
top_popular_car_brand_with_relative_count()
```

```python
def top_registered():
    return df.groupby(['VIN', 'BRAND']).size().nlargest(5)
```

Car brand with the biggest difference in make year in dataset

```python
def car_brand_with_biggest_diff_in_make_year():
    df_f = df[df['MAKE_YEAR'] >=2005]
    df_min = df_f.groupby(['BRAND'])['MAKE_YEAR'].min().reset_index(name="min")
    df_max = df_f.groupby(['BRAND'])['MAKE_YEAR'].max().reset_index(name="max")
    df_merge = pd.merge(df_min, df_max, how='inner', on=['BRAND'])
    df_merge['diff'] = df_merge['max'] - df_merge['min']

    return df_merge.sort_values(by=['diff']).head(5)
```

```python
car_brand_with_biggest_diff_in_make_year()
```

```python
Find cars with the biggest difference between make year and registration year
```

```python
def cars_with_biggest_diff_between_make_year_and_registration_year():
    df['date'] = pd.to_datetime(df['D_REG'], format='%d.%m.%Y')
    df['year'] = pd.DatetimeIndex(df['date']).year
    df['month'] = pd.DatetimeIndex(df['date']).month
    df['diff'] = df['year'] - df['MAKE_YEAR']
    return df.loc[df['diff'].idxmax()]
```

```python
cars_with_biggest_diff_between_make_year_and_registration_year()
```
